# treinamento

```python
train_data = Orange.data.Table("new_train.csv")
new_domain = Orange.data.Domain(
    train_data.domain.attributes[:-1],  # Atributos preditores
    train_data.domain.attributes[-1]  # Classe alvo
)
train_data = Orange.data.Table.from_table(new_domain, train_data)

test_data = Orange.data.Table("test.csv")
new_domain = Orange.data.Domain(
    test_data.domain.attributes[:-1],  # Atributos preditores
    test_data.domain.attributes[-1]  # Classe alvo
)

test_data = Orange.data.Table.from_table(new_domain, test_data)

test_data2 = Orange.data.Table("test_sem_id.csv")
new_domain = Orange.data.Domain(
    test_data2.domain.attributes[:-1],  # Atributos preditores
    test_data2.domain.attributes[-1]  # Classe alvo
)

test_data2 = Orange.data.Table.from_table(new_domain, test_data2)
```

```python
cn2_learner = Orange.classification.rules.CN2Learner()
#cn2_learner.rule_finder.min_covered_examples = 800
classifier = cn2_learner(train_data)
```

```python
predicoes = [classifier(inst) for inst in test_data]

df_test = pd.read_csv("test.csv")  # CSV original para referência
df_test["Previsao"] = predicoes

acertos = (df_test["Previsao"] == df_test["Survived"]).sum()
total = len(df_test)
acuracia = acertos / total

print(f"Acurácia na validação: {acuracia:.2f}")

print(f"Acurácia: {accuracy_score(df_test['Survived'], df_test['Previsao']):.2f}")
print(classification_report(df_test["Survived"], df_test["Previsao"]))
```

```python
predicoes = [classifier(inst) for inst in test_data2]

df_test2 = pd.read_csv("test_sem_id.csv")  # CSV original para referência
df_test2["Previsao"] = predicoes

acertos = (df_test2["Previsao"] == df_test2["Survived"]).sum()
total = len(df_test2)
acuracia = acertos / total

print(f"Acurácia na validação: {acuracia:.2f}")

print(f"Acurácia: {accuracy_score(df_test2['Survived'], df_test2['Previsao']):.2f}")
print(classification_report(df_test2["Survived"], df_test2["Previsao"]))
```
